# Telco Churn EDA

This notebook uses the shared preprocessing module so the same cleaning logic is reused by training.

## EDA

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import seaborn as sns

project_src = Path.cwd() if (Path.cwd() / "preprocess.py").exists() else Path.cwd() / "src"
if str(project_src) not in sys.path:
    sys.path.insert(0, str(project_src))

from preprocess import clean_data, load_data

sns.set_theme(style="whitegrid")
```

```python
data = load_data()
data_clean = clean_data(data)

data_clean.head()
```

```python
sns.countplot(data=data_clean, x="Churn", hue="gender")
plt.title("Churn by gender")
plt.show()

print(data_clean[["Churn", "gender"]].value_counts())
```

```python
sns.kdeplot(data=data_clean, x="tenure", hue="Churn", common_norm=False)
plt.title("Tenure distribution by churn")
plt.show()
```

```python
sns.countplot(data=data_clean, x="Contract", hue="Churn")
plt.title("Churn by contract type")
plt.xticks(rotation=20)
plt.show()
```

```python
sns.countplot(data=data_clean, x="TechSupport", hue="Churn")
plt.title("Churn by Tech Support")
plt.show()

print(data_clean[["TechSupport", "Churn"]].value_counts())
```
